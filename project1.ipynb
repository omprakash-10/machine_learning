```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df=pd.read_csv("C:/Users/ompar/OneDrive/Desktop/weather123.csv")
months=df['Month']
high=df['High']      
low=df['Low']                     
rainfall=df['Rainfall']   
table=[rainfall,high,low]
df=pd.DataFrame(table,index=['Rainfall','High','Low'])
plt.figure(figsize=(18,6))
plt.plot(high,'s-b',linewidth="2")   
plt.plot(low,'s-r',linewidth="2") 
plt.plot(rainfall,'-g',linewidth="3") 
plt.fill_between(months,rainfall,color="green",alpha=0.1)
# set axis title
plt.ylabel("temperature")
plt.grid(axis="y")
# set chart title
plt.title("weather of kathmandu in centegrade in 2024")
plt.legend(['Max temp','Min temp','Rainfall in cm'])
plt.xticks([])
plt.table(cellText=table,rowLabels=['rainafll in cm','Temp in C Max','temp in C min'])
plt.show()
plt                                    

           
           
```
